```python
ds=ds.drop('subject',axis=1)
```

```python
x=ds.drop('Activity',axis=1)
y=ds['Activity']
x,y
```

```python
# Scikit-learn (Sklearn) is the most useful and robust library for machine learning in Python. 
# It provides a selection of efficient tools for machine learning 
# sklearn.neighbors provides functionality for unsupervised and supervised neighbors-based learning methods. 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split 
```

```python
# Split arrays or matrices into random train and test subsets
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
x_test,x_train,y_train,y_test
```

```python
# Classifier implementing the k-nearest neighbors vote.
model=KNeighborsClassifier(n_neighbors=3)
#Fit the k-nearest neighbors classifier from the training dataset given.
model.fit(x_train,y_train)
```

```python
#sklearn. metrics module implements several loss, score, and utility functions to measure classification performance.
from sklearn.metrics import accuracy_score
```

```python
# Predict the class labels for the test data.
y_pred=model.predict(x_test)

#Accuracy classification score.
accuracy_score(y_pred,y_test)*100
```

```python
# The sklearn.tree module includes decision tree-based models for classification and regression.
from sklearn.tree import DecisionTreeClassifier
```

```python
# A decision tree classifier.
dmodel=DecisionTreeClassifier()

#Build a decision tree classifier from the training set (x_train, y_train).
dmodel.fit(x_train,y_train)
```

```python
# Calculating accuracy:
y_pred=dmodel.predict(x_test)
accuracy_score(y_pred,y_test)*100
```

```python
#Export a decision tree in DOT format.
from sklearn.tree import export_graphviz
```

```python
# This function generates a GraphViz representation of the decision tree,
# which is then written into `out_file`. Once exported, graphical renderings can be generated
export_graphviz(dmodel,out_file='dtree.dot')
```

```python
# The goal of ensemble methods is to combine the predictions of several base estimators built with a given learning algorithm in order to improve generalizability / robustness over a single estimator.
from sklearn.ensemble import RandomForestClassifier
```

```python
# A random forest classifier.
# A random forest is a meta estimator that fits a number of decision tree
# classifiers on various sub-samples of the dataset
rmodel=RandomForestClassifier()
rmodel.fit(x_train,y_train)
```

```python
# Calculating accuracy:
y_pred=rmodel.predict(x_test)
accuracy_score(y_pred,y_test)*100
```

```python
# The sklearn.svm module includes Support Vector Machine algorithms.
from sklearn.svm import SVC
```

```python
# C-Support Vector Classification.
# The sklearn.svm module includes Support Vector Machine algorithms.
smodel=SVC(kernel='linear')
smodel.fit(x_train,y_train)
```

```python
y_pred=smodel.predict(x_test)
accuracy_score(y_pred,y_test)*100
```

```python
# TensorFlow is an end-to-end open source platform for machine learning.
# Implementation of the Keras API, the high-level API of TensorFlow.
from tensorflow import keras
```

```python
x_train.shape
```

```python
# Replacing categories with numbers
# Replace values given in `to_replace` with `value`.
y_train.replace({'STANDING':0,
 'SITTING':1,
 'LAYING':2,
 'WALKING':3,
 'WALKING_DOWNSTAIRS':4,
 'WALKING_UPSTAIRS':5},inplace=True)
y_test.replace({'STANDING':0,
 'SITTING':1,
 'LAYING':2,
 'WALKING':3,
 'WALKING_DOWNSTAIRS':4,
 'WALKING_UPSTAIRS':5},inplace=True)
y_train,y_test
```

```python
# `Sequential` groups a linear stack of layers into a `tf.keras.Model`.

# `Sequential` provides training and inference features on this model.
ann=keras.Sequential([
    keras.layers.Dense(6,input_shape=(561,),activation='sigmoid')
])

# Configures the model for training.
ann.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
    )


ann.fit(x_train,y_train,epochs=10)
```

```python
# Returns the loss value & metrics values for the model in test mode.
ann.evaluate(x_test,y_test)
```

```python
import numpy as np
```

```python
y=ann.predict(x_test)
ann_y_pred=[]
for i in y:
    i=np.argmax(i)
    ann_y_pred.append(i)
    print(labels[i])
```

```python
activities={'STANDING':0,
 'SITTING':0,
 'LAYING':0,
 'WALKING':0,
 'WALKING_DOWNSTAIRS':0,
 'WALKING_UPSTAIRS':0}
for i in ds['Activity']:
    activities[i]+=1
activities
```

```python
df_class_0 = ds[ds['Activity'] == "STANDING"]
df_class_1 = ds[ds['Activity'] == 'SITTING']
df_class_2 = ds[ds['Activity'] == 'LAYING']
df_class_3 = ds[ds['Activity'] == 'WALKING']
df_class_4 = ds[ds['Activity'] == 'WALKING_DOWNSTAIRS']
df_class_5 = ds[ds['Activity'] == 'WALKING_UPSTAIRS']
```

```python
df_class_1_under = df_class_1.sample(activities['WALKING_DOWNSTAIRS'])
df_class_2_under = df_class_2.sample(activities['WALKING_DOWNSTAIRS'])
df_class_3_under = df_class_3.sample(activities['WALKING_DOWNSTAIRS'])
df_class_4_under = df_class_4.sample(activities['WALKING_DOWNSTAIRS'])


df_test_under = pd.concat([df_class_0_under,df_class_1_under,df_class_2_under,df_class_3_under,df_class_4_under, df_class_5], axis=0)
df_test_under
```

```python
x_train_under,x_test_under,y_train_under,y_test_under=train_test_split(df_test_under.drop('Activity',axis=1),df_test_under['Activity'],random_state=42,test_size=0.2)
x_train_under,x_test_under,y_train_under,y_test_under
```
